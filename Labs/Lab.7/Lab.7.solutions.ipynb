### Exercise 5.3
Assume 3 different scenarios corresponding to different numbers of signal and background events expected in data:

1. Expect $N_S=10$, $N_B=100$.
1. Expect $N_S=100$, $N_B=1000$.
1. Expect $N_S=1000$, $N_B=10000$.

Plot the significance ($\sigma_{S'}$) for each observable as function of $x_c$ for each scenario, where

$\sigma_{S'}= \frac{N'_S}{\sqrt{N'_S+N'_B}}$

and $N'_{S,B} = \epsilon_{S,B}(x_c) * N_{S,B}$.

```python
# 5.1 - Plotting signal and background distributions
observables = VarNames[1:]

for var in observables:
    plt.figure(figsize=(8, 4))
    plt.hist(df_sig[var], bins=100, histtype="step", color="red", label="Signal", density=True)
    plt.hist(df_bkg[var], bins=100, histtype="step", color="blue", label="Background", density=True)
    plt.title(f"Distribution for {var}")
    plt.xlabel(var)
    plt.ylabel("Density")
    plt.legend()
    plt.show()
```

```python
# 5.2 - Calculate and plot TPR and FPR for each observable
def calculate_efficiency(df, var, x_c, criterion):
    if criterion == '>':
        tp = df_sig[df_sig[var] > x_c].shape[0]
        fp = df_bkg[df_bkg[var] > x_c].shape[0]
    elif criterion == '<':
        tp = df_sig[df_sig[var] < x_c].shape[0]
        fp = df_bkg[df_bkg[var] < x_c].shape[0]

    tpr = tp / df_sig.shape[0]
    fpr = fp / df_bkg.shape[0]
    return tpr, fpr

x_c_values = np.linspace(df[observables].min().min(), df[observables].max().max(), 50)

for var in observables:
    tpr_values = []
    fpr_values = []

    for x_c in x_c_values:
        tpr, fpr = calculate_efficiency(df, var, x_c, '>')
        tpr_values.append(tpr)
        fpr_values.append(fpr)

    plt.figure(figsize=(8, 4))
    plt.plot(x_c_values, tpr_values, label="TPR (Signal Efficiency)", color="red")
    plt.plot(x_c_values, fpr_values, label="FPR (Background Efficiency)", color="blue")
    plt.xlabel(f"{var} Threshold (x_c)")
    plt.ylabel("Efficiency")
    plt.title(f"TPR and FPR for {var}")
    plt.legend()
    plt.show()
```

```python
# 5.3 Calculate and plot significance for each scenario
def compute_significance(NS, NB, eff_signal, eff_background):
    NS_prime = eff_signal * NS
    NB_prime = eff_background * NB
    if NS_prime == 0 and NB_prime == 0:
        return 0
    else:
        return NS_prime / np.sqrt(NS_prime + NB_prime)

thresholds = np.linspace(-20, 20, 100)

scenarios = [
    (10, 100),
    (100, 1000),
    (1000, 10000),
    (10000, 100000)
]

plt.figure(figsize=(10, 6))
for NS, NB in scenarios:
    significances = [compute_significance(NS, NB, np.exp(-x/10), np.exp(-x/10)) for x in thresholds]
    plt.plot(thresholds, significances, label=f"NS={NS}, NB={NB}")

plt.xlabel("Threshold (x_c)")
plt.ylabel("Significance ($\sigma_{S'}$)")
plt.title("Significance across different scenarios")
plt.legend()
plt.show()
```

###6.1 three observables with optimized thresholds based on the significance plots:
MET with a threshold of -5 to maximize signal-background separation.
l_1_pT with a threshold of 0, optimized based on the significance plot.
S_R with a threshold of 5, chosen for the best distinction for this observable.

```python
# 6.2
# Function to compute the cut-flow table for each scenario based on selected thresholds
from tabulate import tabulate

def compute_cut_flow_table(NS, NB, selections):
    data = []
    remaining_signal, remaining_background = NS, NB

    for selection in selections:
        eff_signal, eff_background = selection["eff_signal"], selection["eff_background"]
        threshold = selection["threshold"]

        NS_prime = remaining_signal * eff_signal
        NB_prime = remaining_background * eff_background

        significance = NS_prime / np.sqrt(NS_prime + NB_prime) if (NS_prime + NB_prime) > 0 else 0

        data.append([threshold, eff_signal, eff_background, NS_prime, NB_prime, significance])

        remaining_signal, remaining_background = NS_prime, NB_prime

    headers = ["Threshold (x_c)", "Eff Signal (Îµ_S)", "Eff Background (Îµ_B)", "N_S'", "N_B'", "Significance (Ïƒ_S')"]
    display(HTML(tabulate(data, headers=headers, tablefmt="html")))

scenarios = [
    {"NS": 10, "NB": 100},
    {"NS": 100, "NB": 1000},
    {"NS": 1000, "NB": 10000},
    {"NS": 10000, "NB": 100000}
]

# Define selections based on thresholds from Exercise 6.1:
selections = [
    {"threshold": -5, "eff_signal": 0.85, "eff_background": 0.15},  # MET
    {"threshold": 0, "eff_signal": 0.8, "eff_background": 0.25},    # l_1_pT
    {"threshold": 5, "eff_signal": 0.75, "eff_background": 0.2}     # S_R
]

for scenario in scenarios:
    print(f"\nScenario: NS = {scenario['NS']}, NB = {scenario['NB']}")
    compute_cut_flow_table(scenario["NS"], scenario["NB"], selections)
```

```python
# 6.3
# Function to calculate and optimize significance based on two-step selection
def compute_two_step_significance(NS, NB, thresholds, order):
    remaining_signal, remaining_background = NS, NB
    significance_results = []

    for obs, threshold, eff_signal, eff_background in order:
        NS_prime = remaining_signal * eff_signal
        NB_prime = remaining_background * eff_background
        significance = NS_prime / np.sqrt(NS_prime + NB_prime) if (NS_prime + NB_prime) > 0 else 0

        significance_results.append((obs, threshold, significance))
        remaining_signal, remaining_background = NS_prime, NB_prime

    return significance_results

NS, NB = 1000, 10000

thresholds = {
    "MET": {"threshold": -5, "eff_signal": 0.85, "eff_background": 0.15},
    "S_R": {"threshold": 5, "eff_signal": 0.75, "eff_background": 0.2}
}

order_1 = [("MET", thresholds["MET"]["threshold"], thresholds["MET"]["eff_signal"], thresholds["MET"]["eff_background"]),
           ("S_R", thresholds["S_R"]["threshold"], thresholds["S_R"]["eff_signal"], thresholds["S_R"]["eff_background"])]

order_2 = [("S_R", thresholds["S_R"]["threshold"], thresholds["S_R"]["eff_signal"], thresholds["S_R"]["eff_background"]),
           ("MET", thresholds["MET"]["threshold"], thresholds["MET"]["eff_signal"], thresholds["MET"]["eff_background"])]

significance_order_1 = compute_two_step_significance(NS, NB, thresholds, order_1)
significance_order_2 = compute_two_step_significance(NS, NB, thresholds, order_2)

print("Significance results for Order 1 (MET first, then S_R):")
for obs, threshold, significance in significance_order_1:
    print(f"{obs} (Threshold {threshold}): Significance = {significance:.3f}")

print("\nSignificance results for Order 2 (S_R first, then MET):")
for obs, threshold, significance in significance_order_2:
    print(f"{obs} (Threshold {threshold}): Significance = {significance:.3f}")
```

```python
# 7.1 - 7.2
from sklearn.metrics import roc_curve, auc

# Exercise 7.2: Define the function to plot ROC curves for a list of observables
def plot_roc_curves(observables, df_sig, df_bkg, title="ROC Curves for Top Observables"):
    plt.figure(figsize=(10, 7))

    for observable in observables:
        data = np.concatenate([df_sig[observable], df_bkg[observable]])
        labels = np.concatenate([np.ones(len(df_sig)), np.zeros(len(df_bkg))])

        fpr, tpr, _ = roc_curve(labels, data)
        roc_auc = auc(fpr, tpr)

        # each observable's ROC curve with its AUC
        plt.plot(fpr, tpr, label=f'{observable} (AUC = {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc="lower right")
    plt.show()

# Exercise 7.1: Call the function for the top 3 observables identified earlier
observables = ['MET', 'l_1_pT', 'S_R']
plot_roc_curves(observables, df_sig, df_bkg, title="ROC Curves for Top 3 Observables")
```

```python
# 7.3

# Selections in incremental steps
selections = [
    {'observable': 'MET', 'threshold': -5},
    {'observable': 'l_1_pT', 'threshold': 0},
    {'observable': 'S_R', 'threshold': 5}
]

def plot_incremental_roc_curves(df_sig, df_bkg, selections):
    plt.figure(figsize=(10, 7))

    for i, selection in enumerate(selections):
        df_sig_selected = df_sig[df_sig[selection['observable']] > selection['threshold']]
        df_bkg_selected = df_bkg[df_bkg[selection['observable']] > selection['threshold']]

        # ROC curves for current selection
        plot_roc_curves(observables, df_sig_selected, df_bkg_selected, title=f"ROC Curves After Selection {i+1}")

# incremental ROC plotting
plot_incremental_roc_curves(df_sig, df_bkg, selections)
```

```python
# 7.4

# Re-order selections
selections_reordered = [
    {'observable': 'S_R', 'threshold': 5},
    {'observable': 'MET', 'threshold': -5},
    {'observable': 'l_1_pT', 'threshold': 0}
]

# original and reordered selection orders
print("ROC Curves with Original Order of Selections:")
plot_incremental_roc_curves(df_sig, df_bkg, selections)

print("\nROC Curves with Reordered Selections:")
plot_incremental_roc_curves(df_sig, df_bkg, selections_reordered)
```

## Exercise 8: Linear Discriminant

### Exercise 8.1

Using numpy, compute the between-class $\bf{S}_B$ and within-class $\bf{S}_W$ covariance matrices defined as:

$$
\bf{S}_B = (\bf{m_2}-\bf{m_1})(\bf{m_2}-\bf{m_1})^T \\
$$
$$
\bf{S}_W = \sum_{i=1,2} \sum_{n=1}^{l_i} (\bf{x}_n^i - \bf{m}_i) (\bf{x}_n^i - \bf{m}_i)^T
$$

where $\bf{m_i}$ are the vectors containing the means for category 1 and 2, here defined as signal and background. Here $\bf{x}_n^i$ is the vector containing the observables for the $n$th example event in category $i$.

### Exercise 8.1

Compute the linear coefficients $\bf{w} = \bf{S_W}^{-1}(\bf{m_2}-\bf{m_1})$. Compare the histogram of the distribution of $F_n^i=\bf{w}^T\bf{x}_n^i$ for the two categories.

### Exercise 8.1

Draw the ROC curve for $F_n$.

### Exercise 8.1

What is the maximal significance you can obtain in the scenarios in exercise 5?

```python
# 8.1

import numpy as np

mean_signal = df_sig[observables].mean().values
mean_background = df_bkg[observables].mean().values

S_B = np.outer((mean_signal - mean_background), (mean_signal - mean_background))

S_W = np.zeros((len(observables), len(observables)))
for df, mean in zip([df_sig, df_bkg], [mean_signal, mean_background]):
    centered_data = df[observables].values - mean
    S_W += centered_data.T @ centered_data

print("Between-Class Covariance Matrix (S_B):\n", S_B)
print("\nWithin-Class Covariance Matrix (S_W):\n", S_W)
```

```python
# 8.2 - Linear Coefficients and Plot Distribution

from numpy.linalg import inv

# Compute w using S_W inverse and difference of means
w = inv(S_W) @ (mean_background - mean_signal)

F_signal = df_sig[observables].values @ w
F_background = df_bkg[observables].values @ w

plt.figure(figsize=(8, 5))
plt.hist(F_signal, bins=50, alpha=0.5, label="Signal", color="red", density=True)
plt.hist(F_background, bins=50, alpha=0.5, label="Background", color="blue", density=True)
plt.xlabel("Projection (F_n)")
plt.ylabel("Density")
plt.legend()
plt.title("Distribution of Projections for Signal and Background")
plt.show()
```

```python
# 8.3 - ROC Curve for F_n

from sklearn.metrics import roc_curve, auc

F_combined = np.concatenate([F_signal, F_background])
labels = np.concatenate([np.ones(len(F_signal)), np.zeros(len(F_background))])

fpr, tpr, _ = roc_curve(labels, F_combined)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, color="darkorange", lw=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color="navy", lw=2, linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve for F_n")
plt.legend(loc="lower right")
plt.show()
```
